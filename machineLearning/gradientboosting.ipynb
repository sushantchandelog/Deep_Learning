```python
#traing the Gradient boosting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
housing = fetch_california_housing()
X = pd.DataFrame(housing.data, columns=housing.feature_names)
y = pd.Series(housing.target, name="MedHouseVal")
```

```python
print(X.shape)
print(X.head())
```

```python
#so tree based model i do not need to standardize trhe data
#creating a gradient boost
tree_reg_1 = DecisionTreeRegressor(max_depth=2)
tree_reg_1.fit(X,y)
ypred1 = tree_reg_1.predict(X)
```

```python
#traint the next tree on the residual of the 1st
y2 = y - ypred1
tree_reg_2 = DecisionTreeRegressor(max_depth=2)
tree_reg_2.fit(X,y2)
ypred2 = tree_reg_2.predict(X)
```

```python
y3 = y2 - ypred2
tree_reg_3 = DecisionTreeRegressor(max_depth=2)
tree_reg_3.fit(X,y3)
ypred3 = tree_reg_3.predict(X)
```

```python
y_pred = sum(tree.predict(X) for tree in (tree_reg_1, tree_reg_2, tree_reg_3))

print(y_pred)
```

```python
#creating a Gradient Boosting

gbrt = GradientBoostingRegressor(
    max_depth = 2,
    n_estimators=120
)
gbrt.fit(X,y)
```

```python
from sklearn.metrics import mean_squared_error
errors = [mean_squared_error(y, y_pred) for y_pred in gbrt.staged_predict(X)]
best_est = np.argmin(errors)
```

```python
print(best_est)
```

```python
gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators=best_est)
gbrt_best.fit(X,y)
```

```python
y_hat = gbrt_best.predict(X)
print(y_hat)
```

```python
print("Error", mean_squared_error(y,y_hat))
```
