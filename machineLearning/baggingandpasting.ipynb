```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import fetch_openml

```

```python
housing = fetch_openml(name="house_prices")

X = housing.data
y = housing.target
```

```python
print(X.shape  ,"/", y.shape)
```

```python
#the data is so messy
X = pd.DataFrame(X)
y = pd.Series(y)
```

```python
print(X.info())
print("-"*60)
```

```python
print(X.isna().sum().to_string())
```

```python
list_drop = X.columns[X.isnull().any()]
```

```python
X = X.drop(list_drop, axis=1)
```

```python
X.shape
```

```python
X = pd.get_dummies(X)
```

```python
bag_clf = BaggingClassifier(

    DecisionTreeClassifier(), n_estimators=50,

    max_samples=100, bootstrap=True, n_jobs=-1

)

bag_clf.fit(X,y)
```

```python
ypred = bag_clf.predict(X)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
print(accuracy_score(y, ypred))
```
