```python
#Minst Dataset
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

```python
X , y = mnist["data"], mnist["target"]
print(X.shape)
print(y.shape)
X = X.to_numpy()
y = y.to_numpy()
```
